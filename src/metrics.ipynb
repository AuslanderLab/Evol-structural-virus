```python
import pandas as pd
import numpy as np
```

```python
def jaccard(df, outpath, mat_outpath=None):
    """
    Calculating pairwise jaccard index
    :param filepath: path to the taxonomy counts
    :param type: str
    :param outpath: path to write the similarity and species names to
    :param type: str
    :param mat_outpath: path to write in matrix form
    :param type: str
    """
    df=pd.read_csv(df, index_col="cluster_ID")
    # data is a list of nexted lists used to construct the dataframe post-loop
    data=[]
    # df column names
    data_names=["taxon1", "taxon2", "intersection", "union", "jaccard index"]
    # iterating over all specie
    for col_i in df.columns:
        # turn 0 and 1 into False/True
        col1=df[col_i].astype(bool)
        for col_j in df.columns:
            # turn 0 and 1 into False/True
            col2=df[col_j].astype(bool)
            # sum if col1 or col1 is 1
            union=np.sum((col1 | col2))
            # must be present in col1 and 2
            intersection=np.sum((col1 &col2))
            # the jaccard index
            jaccard=intersection/union
            # adding a new entry into the data list
            data.append([col_i,col_j, intersection, union, jaccard])
    # turn the nested list into a dataframe
    df=pd.DataFrame(data,columns=data_names)
    # write df to path
    df.to_csv(outpath, index=False)
    # select the taxons
    df=df[["taxon1","taxon2","jaccard index"]]
    # pivot to sparse matrix
    df=df.pivot(index="taxon1", columns="taxon2", values="jaccard index")
    df=df.fillna(0)
    # write sparse matrix to file
    df.to_csv(mat_outpath,index=True)
```

```python
def main():
    ### THIS IS THE CALCULATION FROM THE STRUCTURE-BASED CLUSTERING
    counts="../intermediate_files/structure_df.csv"
    output="../results/species_jaccard.csv"
    jaccard(counts, output, mat_outpath=None)

    ### THIS is the AMINO ACID PROTEIN CLUSTERS FROM MMSEQS2
    amino_acid_count_df="../intermediate_files/amino_acid_count_df.csv"
    output_jaccard_df="../results/amino_acid_jaccard_distance.csv"
    jaccard(amino_acid_count_df, output_jaccard_df, mat_outpath=None)

    ## This is for the high quality structure predictions
    jaccard("../intermediate_files/high_quality_structure_results/plddt_0_count.csv",
            "../results/plddt_0_jaccard.csv")
    jaccard("../intermediate_files/high_quality_structure_results/plddt_50_count.csv",
            "../results/plddt_50_jaccard.csv")
    jaccard("../intermediate_files/high_quality_structure_results/plddt_70_count.csv",
            "../results/plddt_70_jaccard.csv")
    jaccard("../intermediate_files/high_quality_structure_results/plddt_90_count.csv",
            "../results/plddt_90_jaccard.csv")
    


if __name__=="__main__":
    main()
```
